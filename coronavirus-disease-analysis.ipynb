<h4>2. Write a python program to replace all the missing values with
“Tobefound” for province column.</h4>

```python
df['Province_State'].replace(np.nan, 'Tobefound', inplace=True)
```

<h4>3. Normalize the values for “Confirmed” column as per the z-score.</h4>

```python
mean_confirmed = df['Confirmed'].mean()
std_confirmed = df['Confirmed'].std()
df['Confirmed_zscore'] = (df['Confirmed'] - mean_confirmed) / std_confirmed
df.head()
```

<h4>4. Create bin values for “Death” column in the data set.</h4>

```python
group_names = ['Low', 'Medium', 'High']
bins = np.linspace(min(df['Deaths']), max(df['Deaths']), 4)
df['Deaths_Binned'] = pd.cut(df['Deaths'], bins, labels=group_names, include_lowest=True)
df[['Deaths','Deaths_Binned']].head(20)
```

<h4>5. Write a Python program to get the maximum number of confirmed,
deaths, recovered and active cases of Novel Coronavirus (COVID-19)
Country wise.</h4>

```python
print(df['Active'].isna().sum())  # Count of NaN values
print(df['Active'].describe())   # Summary of column statistics
```

The output confirms that the Active column contains only NaN values. To resolve this, we calculate the "Active" values.

```python
df['Active'] = df['Confirmed'] - (df['Deaths'] + df['Recovered'])
```

```python
print(
    f"Maximum number of confirmed is {df['Confirmed'].max()} in {df.loc[df['Confirmed'].idxmax(skipna=True), 'Country_Region']}.\n"
    f"Maximum number of deaths is {df['Deaths'].max()} in {df.loc[df['Deaths'].idxmax(skipna=True), 'Country_Region']}.\n"
    f"Maximum number of recovered is {df['Recovered'].max()} in {df.loc[df['Recovered'].idxmax(skipna=True), 'Country_Region']}.\n"
    f"Maximum number of active cases of Novel Coronavirus (COVID-19) is {df['Active'].max()} in {df.loc[df['Active'].idxmax(skipna=True), 'Country_Region']}."
)
```

<h4>6. Write a Python program to get the total number of confirmed deaths
and recovered people of Novel Coronavirus (COVID-19) cases
Country/Region - Province/State wise</h4>

```python
# Group by Country/Region and Province/State
summary = df.groupby(['Country_Region', 'Province_State'])[['Confirmed', 'Deaths', 'Recovered']].sum()

# Reset the index for a cleaner DataFrame
summary = summary.reset_index()
summary
```

<h4>7. Write a Python program to get the country wise (country name taken as
input from user) cases of confirmed, deaths and recovered cases of
Novel Coronavirus (COVID-19).</h4>

```python
# Get user input for the country name
country = input("Enter the country name: ")

# Filter the DataFrame for the selected country
country_data = df[df['Country_Region'].str.contains(country, case=False, na=False)]

# Summarize confirmed, deaths, and recovered cases
summary = country_data[['Confirmed', 'Deaths', 'Recovered']].sum()

print(f"\nSummary for {country}:")
print(f"Total Confirmed Cases: {summary['Confirmed']}")
print(f"Total Deaths: {summary['Deaths']}")
print(f"Total Recovered Cases: {summary['Recovered']}")
```
